```python
import pandas as pd
```

```python
orders=pd.read_csv(r"D:\ai_customer_intelligence\data\olist_orders_dataset.csv")
items=pd.read_csv(r"D:\ai_customer_intelligence\data\olist_order_items_dataset.csv")
products=pd.read_csv(r"D:\ai_customer_intelligence\data\olist_products_dataset.csv")
reviews=pd.read_csv(r"D:\ai_customer_intelligence\data\cleaned_reviews_dataset.csv")
translation=pd.read_csv(r"D:\ai_customer_intelligence\data\product_category_name_translation.csv")
```

```python
orders
```

```python
orders=orders[['order_id','customer_id','order_status','order_purchase_timestamp']]
orders
```

```python
products=products[['product_id','product_category_name']]
products
```

```python
items=items[['order_id','product_id','price']]
items
```

```python
reviews=reviews[['order_id','review_score','review_clean']]
reviews
```

```python
reviews=reviews.rename(columns={'review_score':'score','review_clean':'comment'})
```

```python
reviews.isna().sum()
```

```python
reviews=reviews.dropna(subset=['comment'])
```

```python
reviews.shape
```

```python
translation
```

```python
products=products.merge(translation,on='product_category_name',how='left')
products
```

```python
master=orders.merge(items,on='order_id').merge(products,on='product_id').merge(reviews,on='order_id',how='left')
master
```

```python
master=master[master['order_status']=='delivered']
master=master.dropna(subset=['product_category_name_english'])
master['comment']=master['comment'].fillna('sem comentários')
```

```python
master
```

```python
master['order_purchase_timestamp']=pd.to_datetime(master['order_purchase_timestamp'])
```

```python
master['month']=master['order_purchase_timestamp'].dt.month
master['year']=master['order_purchase_timestamp'].dt.year
master['year_month']=master['order_purchase_timestamp'].dt.to_period('M')
```

```python
master.to_csv(r'D:\ai_customer_intelligence\data\master_table_dataset.csv',index=False)
```
