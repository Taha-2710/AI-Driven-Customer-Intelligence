```python
import pandas as pd
df=pd.read_csv(r"D:\ai_customer_intelligence\data\olist_order_reviews_dataset.csv")
```

```python
df=df[['order_id','review_score','review_comment_message']]
```

```python
df=df.dropna(subset=['review_comment_message'])
```

```python
df
```

```python
df['review_clean'] = df['review_comment_message'].str.lower()
```

```python
df
```

```python
import re

def clean_review(text):
    text = text.strip()  # remove extra spaces
    
    # Check if text contains at least one alphabet (Portuguese included)
    if re.search(r'[a-zà-ú]', text):
        return text
    else:
        return None
```

```python
df['review_clean'] = df['review_clean'].apply(clean_review)
```

```python
df = df[df['review_clean'].notna()]
```

```python
df
```

```python
print(df[['review_comment_message', 'review_clean']])
```

```python
def remove_meaningless_token(text):
    words=text.split()
    cleaned_words=[]
    for word in words:
        if len(word)==1:  # remove single character
            continue
        if len(set(word))==1: # remove words with only one unique character
            continue
        cleaned_words.append(word)
    return " ".join(cleaned_words)

df['review_clean']=df['review_clean'].apply(remove_meaningless_token)
```

```python
df
```

```python
import re

def normalize_elongated_words(text):
    # reduce repeated characters to max 2 occurrences
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    return text
```

```python
df['review_clean']=df['review_clean'].apply(normalize_elongated_words)
```
