```python
import pandas as pd
master=pd.read_csv('master_table_dataset.csv')

reviews=pd.read_csv('reviews_with_sentiment_dataset.csv')
master['score']=master['score'].fillna(0)
master.isna().sum()
master_sentiment=master.merge(reviews[['order_id','sentiment_score','sentiment']],on='order_id',how='left')

master_sentiment['sentiment_score']=master_sentiment['sentiment_score'].fillna(0)
master_sentiment['sentiment']=master_sentiment['sentiment'].fillna('Sem comentários')

complaints=pd.read_csv('classified_complaints.csv')
print(complaints.columns)
master_final=master_sentiment.merge(complaints[['order_id','complaint_type','confidence_score']],on='order_id',how='left')

master_final.shape
master_sentiment.shape
master_final['confidence_score']=master_final['confidence_score'].fillna(0)
master_final['complaint_type']=master_final['complaint_type'].fillna('Sem comentários')

master_final=master_final[['order_id','product_category_name','price','score','sentiment_score','sentiment','complaint_type','confidence_score','year_month','month','year']]
master_final.to_csv('master_final.csv',index=False)
```
