```python
import pandas as pd
master_final=pd.read_csv('master_final.csv')

from numpy import average
sales_sentiment=master_final.groupby(['year_month','product_category_name']).agg(total_revenue=('price','sum'),
                                                                               order_count=('order_id','count'),
                                                                               avg_review_score=('score','mean'),
                                                                               average_sentiment_score=('sentiment_score','mean'),
                                                                               negative_count=('sentiment',lambda x: (x=='Negative').sum()),
                                                                               positive_count=('sentiment',lambda x: (x=='Positive').sum())).reset_index()
sales_sentiment.to_csv('sales_sentiment.csv',index=False)
```
